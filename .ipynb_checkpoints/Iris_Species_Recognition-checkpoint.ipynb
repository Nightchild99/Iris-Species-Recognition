FELADAT: A megadott dataset alapján feltanítani egy neurális hálót tensorflow segítségével.  
A dataset-ben a 'Appliances' oszlopot kell majd megprediktálni.

Forrás állomány: https://www.kaggle.com/datasets/loveall/appliances-energy-prediction/data?select=KAG_energydata_complete.csv

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout
```

### Töltsük be a megadott KAG_energydata_complete.csv-t és irassuk ki az első 10 sorát.

```python
# Adatbetöltés
df = pd.read_csv("KAG_energydata_complete.csv")
df.head()
```

### Készítsük elő az adatot a tanításhoz. Vizsgáljuk meg, hogy van-e hiányzó adatunk, ha van akkor kezeld őket. Dobjuk el azokat az oszlopokat amiket nem fogunk tudni használni.

```python
# Adat-előkészítés
missing_values = df.isnull().sum()
missing_values
# Hiányzó értékek meghatározása az adathalmazban

# Oszlopok eltávolítása
df = df.drop('date', axis=1)
df.head()
```

### Normalizáljuk az adatokat MinMaxScaler segítségével. Majd váltogassuk szét őket feature és target változókra.

```python
# Adatok normalizálása MinMaxScalerrel
scaler = MinMaxScaler()
df_scaled = scaler.fit_transform(df)
df_scaled = pd.DataFrame(df_scaled, columns=df.columns)
df = df_scaled
# Targer és feautre létrehozása
target = df['Appliances']
features = df.drop('Appliances', axis=1)
```

### Válasszuk szét az adatokat train és test adathalmazra. Test size legyen 20%, random state pedig 42

```python
# Adatkészlet felosztása tanító és teszt halmazra
X_train, X_test, y_train, y_test = train_test_split(features, target, test_size=0.2, random_state=42)
(X_train.shape, X_test.shape, y_train.shape, y_test.shape)
```

### Építsünk fel egy egyszerű modellt tensorflow segítségével
A következő dense rétegek add hozzá a modellhez:
- 128 Dense, activation = relu, add meg a bemeneti dimenziót
- 64 Dense, activation = relu
- 32 Dense, activation = relu
- 1 Dense, activation = linear

```python
# Modell építése
model = Sequential([
    Dense(128, activation='relu', input_shape=(27,)),
    Dropout(0.2),
    Dense(64, activation='relu'),
    Dropout(0.2),
    Dense(32, activation='relu'),
    Dropout(0.2),
    Dense(1, activation='linear')
])

# Modell összeállítása
model.compile(optimizer='adam', loss='mean_squared_error', metrics=['accuracy'])

model.summary()
```

### Tanítsuk fel a modeellt. Epoch száma legyen 100 és a batch_size-unk pedig 32.


```python
# Modell tanítása
history = model.fit(X_train, y_train, validation_split=0.2, batch_size=32, epochs=100, verbose=1)
```

### Készítsünk egy plotot, amin ábrázoljuk a loss-t és a validation-lost az epochok alatt. Labelezd fel és adj címet is a plot-nak.

```python

loss = history.history['loss']
val_loss = history.history['val_loss']
epochs = range(1, len(loss) + 1)

# Veszteség grafikon
plt.figure(figsize=(14, 5))
plt.subplot(1, 2, 1)
plt.plot(epochs, loss, 'bo', label='Tanítási veszteség')
plt.plot(epochs, val_loss, 'b', label='Validációs veszteség')
plt.title('Tanítási és Validációs Veszteség')
plt.xlabel('Epochok')
plt.ylabel('Veszteség')
plt.legend()
plt.show()
```

```python
# Modell értékelése
test_loss = model.evaluate(X_test, y_test)
print(f'Test Loss: {test_loss}')
```
